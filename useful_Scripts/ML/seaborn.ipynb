# Distribution plot
    1.distplot
    2.joinplot
    3.pairplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
#joint Plot
#univarient Analysis -- helps to do univarient analysis
sns.jointplot(x ="tip",y = "total_bill",data = df,kind = "hex")
```

```python
sns.jointplot(x ="tip",y = "total_bill",data = df,kind = "reg")
```

```python
#pair plot --Scater plot
sns.pairplot(df)
```

```python
df.smoker.value_counts()
```

```python
sns.pairplot(df,hue = 'sex')
```

```python
# DISPLOT  --heps to crate histogram
sns.distplot(df.tip)
```

```python
sns.distplot(df.tip,kde = False,bins =10)
```

# categorical plots
    1.boxplot
    2.violinplot
    3.countplot
    4.barplot

```python
#countplot -- Show the counts of observations in each categorical bin using bars.
# sns.countplot(df.sex,data =df)
sns.countplot('sex',data =df)
```

```python
sns.countplot('day',data =df)
```

```python
sns.countplot('smoker',data =df)
```

```python
#changing the axis
sns.countplot(y = 'sex',data =df)
```

```python
#BARPLOT --Show point estimates and confidence intervals as rectangular bars.
#have to give both x and y value
sns.barplot(x= 'day',y = 'total_bill',data = df)
```

```python
#BOXPLOT --Draw a box plot to show distributions with respect to categories.
sns.boxplot('day','total_bill',data =df,hue = 'smoker')
```

```python
#VIOLIN PLOT -- Draw a combination of boxplot and kernel density estimate.

#combination of boxplot and kernal density estimation
```

```python
sns.violinplot(x ='total_bill',y = 'day',data = df,palette = "rainbow")
```
